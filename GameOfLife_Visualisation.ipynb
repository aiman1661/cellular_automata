## Game of Life Animation

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import matplotlib
from GameOfLife_class import GoL_Lattice
matplotlib.use('TKAgg')

import pandas
from pandas import DataFrame as df
pandas.set_option('display.max_rows', 500) # show full data
```

```python
# fixed parameters
iterations = 500

# random initialisation
N = 50
lattice = random.randint(low=0,high=2,size=(N,N))

# set system
system = GoL_Lattice(lattice)

# glider system
#'''
lattice = system.set_glider()
system = GoL_Lattice(lattice)
#'''

# blinker system
'''
lattice = system.set_blinker()
system = GoL_Lattice(lattice)
'''

# beehive system
'''
lattice = system.set_beehive()
system = GoL_Lattice(lattice)
'''

# flower system
'''
lattice = system.set_flower()
system = GoL_Lattice(lattice)
'''

# crab system
'''
lattice = system.set_crab()
system = GoL_Lattice(lattice)
'''

# test system
'''
lattice = system.set_test()
system = GoL_Lattice(lattice)
'''
```

```python
# for any initialisation

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

pop_history = []
generations = []

im = ax1.imshow(system.lattice, animated=True, vmin=-1, vmax=1)
line, = ax2.plot([], [])  # Empty line plot to start

for i in range(iterations):
    lattice_new = system.update_step()
    system = GoL_Lattice(lattice_new)
    
    live_sites = system.count_live()
    pop_history.append(live_sites)
    generations.append(i)

    ax1.cla()
    im=ax1.imshow(system.lattice, animated=True, vmin=-1, vmax=1)
    ax1.set_title('Game of Life Animation')

    # Update sites plot
    line.set_data(generations, pop_history)
    ax2.relim()  # Recalculate limits
    ax2.autoscale_view()  # Autoscale the view
    ax2.set_title('Live sites population over time')
    ax2.set_xlabel('Iterations')
    ax2.set_ylabel('Number of Live Sites')

    plt.draw()
    plt.pause(0.0001)
```

```python
# for glider

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

com_x_array = []
com_y_array = []
time_array = []

im = ax1.imshow(system.lattice, animated=True, vmin=-1, vmax=1)
line, = ax2.plot([], [])  # Empty line plot to start

for i in range(iterations):
    lattice_new = system.update_step()
    system = GoL_Lattice(lattice_new)
    
    com_glider = system.return_com()
    if com_glider is not None and len(com_glider) == 2:
        com_x_array.append(com_glider[0])
        com_y_array.append(com_glider[1])
        time_array.append(i)

    ax1.cla()
    im=ax1.imshow(system.lattice, animated=True, vmin=-1, vmax=1)
    ax1.set_title('Game of Life Animation')

    # update sites plot
    ax2.cla()
    ax2.plot(time_array, com_x_array, 'rx', label='x', markersize=2)
    ax2.plot(time_array, com_y_array, 'gx', label='y', markersize=2)
    ax2.legend()
    ax2.relim()  # Recalculate limits
    ax2.autoscale_view()  # Autoscale the view
    ax2.set_title('Glider CoM Motion')
    ax2.set_xlabel('Time')
    ax2.set_ylabel('Position')

    plt.draw()
    plt.pause(0.0001)
```

```python
# calculate com glider velocity

print(com_x_array)
print(com_y_array)
print(time_array)
```
