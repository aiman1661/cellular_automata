## Game of Life Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TKAgg')

%matplotlib inline
```

```python
data = np.load('gol_eq_search.npy', allow_pickle=True)
```

```python
data_item = data.item()

eq_steps = data_item['Equilibrium_Steps']
sim_data = data_item['Simulation_data']

print(f'Observables : {data_item.keys()}')
print(f'Equilibrium Data : {eq_steps}')
print(f'Full Simulation Data : {sim_data}')
```

```python
style = {'facecolor': 'none', 'edgecolor': 'C0', 'linewidth': 3}

xbins = np.array([i*200 for i in range(36)])

fig, ax = plt.subplots()
ax.hist(eq_steps, bins=xbins, **style)

# plot the xdata locations on the x axis:
ax.plot(eq_steps, 0*eq_steps, 'd')
ax.set_title('Number of times to equilibriate')
ax.set_ylabel('Number of equilibrium steps per bin')
ax.set_xlabel('Timestep Bins')
```
