## SIRS Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TKAgg')

%matplotlib inline
```

```python
data = np.load('SIRS_fix_measurements.npy', allow_pickle=True)
```

```python
data_item = data.item()

I_average = data_item['I_average']
I_squared_average = data_item['I_squared_average']
I_variance = data_item['I_variance']

print(f'observables : {data_item.keys()}')
print(f'Average Infection (fraction) : {I_average}')
print(f'Average Infection (fraction) Squared : {I_squared_average}')
print(f'Variance : {I_variance}')

I_average[8][8]
```

```python
plt.imshow(I_average, cmap='viridis', interpolation='nearest', extent=[0, 1, 1, 0]) 

# Add colorbar 
plt.colorbar() 

plt.xlabel('p3')
plt.ylabel('p1')  
plt.title("Average Infection (fraction)") 
plt.show() 
```

```python
plt.imshow(I_variance, cmap='viridis', interpolation='nearest', extent=[0, 1, 1, 0]) 

# Add colorbar 
plt.colorbar() 

plt.xlabel('p3')
plt.ylabel('p1')  
plt.title("Infection Variance") 
plt.show() 
```

```python
data_var = np.load("var_waves.npy", allow_pickle=True)
```

```python
data_var_item = data_var.item()

I_average = data_var_item['I_average']
I_variance = data_var_item['I_variance']
sigma_var = data_var_item['Sigma_var']
p1 = data_var_item['p1_values']

print(f'Observables : {data_var_item.keys()}')
print(f'p1 : {p1}')
print(f'Average Infection (fraction) : {I_average}')
print(f'Variance : {I_variance}')
print(f'Sigma Variance : {sigma_var}')
```

```python
# First plot (top)
plt.errorbar(p1, I_variance, yerr= sigma_var, fmt='o', ecolor='black', capsize=5, color="blue")
plt.title('Variance plot for slice at constant p3')
plt.xlabel('p1')
plt.ylabel('Variance')
plt.tight_layout()
plt.show()
```

```python
data_immune = np.load('immune_measurements.npy', allow_pickle=True)
```

```python
data_immune_item = data_immune.item()

I_average = data_immune_item['I_average']
im_frac = data_immune_item['im_frac']

print(f'Observables : {data_immune_item.keys()}')
print(f'Immune Fraction : {im_frac}')
print(f'Average Infection (fraction) : {I_average}')
```

```python
# First plot (top)
plt.scatter(im_frac, I_average, marker='x', )
plt.title('Infection (fraction) Average against Immune Fraction')
plt.xlabel('Immune Fraction $f_{im}$')
plt.ylabel('Infection (fraction) Average')
plt.tight_layout()
plt.show()
```
